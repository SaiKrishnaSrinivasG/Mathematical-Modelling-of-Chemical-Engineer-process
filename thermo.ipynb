## Eqauation of state is a thermodynamic  state variable

Equation of state (EOS)is a equation which relates the state variables which can describe any state of matter under given set of physical conditions such as temperature , Pressure , volume .EOS is generally used to describe the properties of pure substances and mixtures in liquid, solid and gas phase .



The simplest known EOS is the ideal gas equation, which we had used extensively from our childhood.
The ideal gas equation tells us that the 4 varibales can not be chnaged independtely , if we know the 3 variables , we can estimate the 4th one .

It is defined by $$ pV = nRT $$
Let $V$ denote the volume of a gas in Litres

Let $T$ denote the Temperature of the gas in 

Let $R$ denote the Universal Gas Constant of a gas.

Let $P$ denote the Pressure  of the gas.


The problem with the EOS is there is no single EOS that predicts the behavior of all substances under all conditions.



There are various EOS used

# Peng–Robinson equation PR

### The Peng–Robinson equation of state (PR EOS) was developed in 1976 at The University of Alberta by Ding-Yu Peng and Donald Robinson 

<div>in order to satisfy the following goals:</div>
<ol>
<li>The parameters should be expressible in terms of the critical properties and the acentric factor.</li>
<li>The model should provide reasonable accuracy near the critical point, particularly for calculations of the compressibility factor and liquid density.</li>
<li>The mixing rules should not employ more than a single binary interaction parameter, which should be independent of temperature, pressure, and composition.</li>
<li>The equation should be applicable to all calculations of all fluid properties in natural gas processes.</li>
</ol>

pr equation.png

Peng Robinson EOS function is used directly from the thermo library , in this case I am trying to estimate the pressures for the compounds mentioned above and using the thermo library , the pressure using the Peng Robsinson EOS for each of the compound
From calculating the $V$, $T$,$P_c$,$T_c$,$w$ we can estimate the pressure through peng robinson EOS

````python
from thermo.chemical import Chemical


def Peng_Robinson_const(X):
    r = 8.314 #   # universal gas constant,j/mol*k
    tt = Chemical(X) # taking the compounds
    p = tt.P/100000            #pressure
    v = tt.Vml_STP
    t = tt.T
    pc = tt.Pc/100000.  # critical pressure
    tc = tt.Tc       # critical temperature

    w = -np.log10(p/pc)-1.0 #accentric factor 

    k = 0.37464 + (1.54226 * w) - (0.26992 * (w**2))
    
    #a =  (1+(k*(1-(np.sqrt(t/tc)))))**2
    a = tt.omega
    b = 0.07780*((r*tc )/   pc)

    atc = 0.045724 * ( ( a*(r**2)*(tc**2)  ) / pc)

    at = a *atc

    result = [r,t,v,b,at]

    #result  = ( (r*t)/ (v-b)) - (  (at) / ( (v*(v+b)) + (b*(v-b)) ) )

    return result
````

````python
def Peng_Robinson_calc(x):

    result  = ( (x[0]*x[1])/ (x[2]-x[3])) - (  (x[4]) / ( (x[2]*(x[2]+x[3])) + (x[3]*(x[2]-x[3])) ) ) 

    return result 

````

````python
for i in compounds:
    print(f'The  Pressure  of {i} is ----->{Peng_Robinson_calc(Peng_Robinson_const(i))}')
````

The results tell us about the pressure for different compounds using the Peng Robinson EOS.

# Redlich–Kwong equation RK

rk equation.png

````python
def Redlich_Kwong_const(X):
    from thermo.chemical import Chemical
    r = 0.08206 # j/mol*k
    tt = Chemical(X)
    v = tt.Vml_STP
    t = tt.T
    pc = tt.Pc/100000
    tc = tt.Tc


    a = 0.42748*((  (r**2)*(tc**2.5)  )/   pc )


    b = 0.08664*((   r*tc     )/pc  ) 

    result = [r, t, v, b, a]
    #result = (   (r*t)  / ( v-b  ) )   -   ( ( a ) /  ( (np.sqrt(t)*v)*(v+b)   )    )  

    return result

    
````

````python
def Redlich_Kwong_calc(x):
    result = (   (x[0]*x[1])  / ( x[2]-x[3]  ) )   -   ( ( x[4] ) /  ( ((x[1]**2)*x[2])*(x[2]+x[3])   )    ) 
    return result
````

````python
for i in compounds:
    print(f'The Redlich Kwong of {i}is ----->{Redlich_Kwong_calc(Redlich_Kwong_const(i))}')
````
